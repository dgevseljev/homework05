## Задание 1  

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy import special
```

```python
x = np.linspace(-1.0, 1.0, 500)
y = x ** 2 * np.sin(1 / x ** 2) + x
plt.plot(x, y)
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('рис. 1.1')
```

```python
x = np.linspace(5, 25, 100)
y = 1 - ((1 / (1 + x ** 2)) / (1 / x ** 2))
plt.semilogy(x, y)
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('рис. 1.2')
```

```python
x = np.linspace(-2, 2, 100) 
plt.plot(x, special.entr(x),color = 'red', label='Elementwise function for computing entropy') 
plt.plot(x, special.erf(x), color = 'green', label='The error function of complex argument' ) 
plt.plot(x, special.kolmogorov(x), color = 'blue', label='Function of Kolmogorov distribution.')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('рис 1.3')
plt.legend()
```

```python
x = np.linspace(-6*np.pi, 6*np.pi, 250)
y = np.sin(x) / x
plt.figure(figsize=(15, 5))
plt.plot(x, y)
points = np.arange(-6 * np.pi, 6.001 * np.pi, np.pi / 2)
plt.xticks(points, list(map(lambda x: x + "π", map(str, np.arange(-6, 6.001, 0.5)))))
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('рис. 2')
plt.grid()
```

## Задание 3


```python
x = np.linspace(-np.pi, np.pi, 500)
y = np.linspace(-np.pi, np.pi, 500)
X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.sin(Y)
fig, ax = plt.subplots(figsize=(6,5))
norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, norm = norm, cmap="prism")
ax.axis('tight')
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(5))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(5))

cb = fig.colorbar(p, ax=ax)
cb.set_ticks([1, -.5, 0, .5, 1])
plt.title('рис. 3.1')
```

```python
x = np.linspace(-3, 3, 500)
y = np.linspace(-3, 3, 500)
X,Y = np.meshgrid(x,y)

Z = np.exp(-X**2 - Y**2)
fig, ax = plt.subplots(figsize=(6,5))
norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, norm = norm, cmap="hot")
ax.axis('tight')
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

cb = fig.colorbar(p, ax=ax)
cb.set_ticks([-1, -.5, 0, .5, 1])
plt.title('рис. 3.2')
```
